# Mise à jour d'un projet Symfony avec Doctrine

## Étape 1 : Vérification de la connexion et création de la base de données

```python

# Vérifiez que Symfony est installé et les dépendances sont à jour
!composer install

# Vérifiez que la connexion à la base de données est correcte dans le fichier .env ou .env.local

# Création de la base de données
!php bin/console doctrine:database:create --if-not-exists
```

## Étape 2 : Validation du schéma de la base de données

```python

# Validez le schéma pour vérifier les incohérences entre les entités et la base de données
!php bin/console doctrine:schema:validate
```

## Étape 3 : Synchronisation du schéma avec les entités

```python

# Synchronisez le schéma avec les entités si nécessaire (sans migrations)
!php bin/console doctrine:schema:update --dump-sql

# Appliquez les changements si tout est correct
!php bin/console doctrine:schema:update --force
```

## Étape 4.0 : Gestion des migrations - si aucune modification

```python

# Ajoute les migrations a la bdd sans les executés.
doctrine:migrations:sync-metadata-storage
doctrine:migrations:version --add --all -n
```

## Étape 4.1 : Gestion des migrations -- si modification effectuer sur une entité après le schema update

```python

# Générez une nouvelle migration pour les modifications non enregistrées
!php bin/console make:migration

# Appliquez toutes les migrations pour synchroniser le schéma
!php bin/console doctrine:migrations:migrate --no-interaction
```

## Étape 5 : Chargement des données initiales (facultatif)

```python

# Si vous avez des fixtures définies, chargez-les dans la base de données
!php bin/console doctrine:fixtures:load --no-interaction
```

## Étape 6 : Vérification finale

```python

# Assurez-vous que tout est à jour
!php bin/console doctrine:schema:validate

# Vérifiez les migrations appliquées
!php bin/console doctrine:migrations:list
```
